## Research Question

Was there a significant difference between crime rates in California college cities/towns compared to the crime rates of their local county in 2024? Does a correlation exists between crime rates and whether a city or town is a college city or town?

## Background and Prior Work

Crime in college and university towns has been widely reported for the last few decades, with government reported statistics going back to the 1990s. The <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Jeanne Clery Disclosure of Campus Security Policy and Campus Crime Statistics Act requires that crime data for universities must be publicly reported through a Department of Education sponsored survey. There is an abundance of data on isolated college crimes, but less so on the crime rates of areas near universities. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)A study in the American Journal of Police compared campus and city crime rates, published in 1995 by Max L. Bromley from the University of South Florida, examined the relationship between the crime rates with the local communities that surround the university campus. The data used in the study relied heavily on the two largest universities in each state and on an average of 2 years for the population. In result, there was a similar widespread variation found when comparing the violent and property crime rates per 1,000 students or city residents. All of the comparisons concluded that the university's crime rates were significantly lower than those of their cities. In addition, there was a finding that property crime was the vast majority of crimes compared to violent crimes, and that the university property crimes were especially higher than those in their cities. Overall, the findings were consistent with the prior studies in the 1980s and the early 1990s. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)A research paper published on November 24 2020 by Kelsey Cundiff at University of Missouri-St. Louis analyzed campus proximity and neighborhood crime rates. It hypothesized with opportunity theory that neighborhoods within close proximity to college campuses will have higher crime rates because “college students are vulnerable to victimisation" due to convergence for their social lifestyles. It used data collected by the National Neighborhood Crime Survey, and analysis was done with a “multilevel negative binomial regression” to analyze the relationship between “proximity… to college campus[es] and rates of violent and property crime”. The researchers doing this study found that “bordering tracts higher rates of larceny, burglary, and robbery,” acting as proxies for “indicators of crime.” <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) A research study published on October 10, 2023 by Carter Prentice and his team conducted a study on whether there was a difference in homicide and suicide rates between college towns/cities vs non college towns. Prentice’s team conducted this study by analyzing county level mortality rates from the CDC death certificate data from the years 2015 to 2019. To determine if a county was deemed a college town they used the classifications of the American Communities Project (ACP) to determine whether it was a college town or not. In college towns, they found that there was a median 2.56 total homicides per 100000 population and in non college towns, a 3.11 total homicide median. But in college towns, they found that people aged less than 24 had a higher rate of firearm homicide compared to non college towns, showing elevated risk in younger people with firearms. The researchers found that the results for population level homicide were not significantly different from college towns and non college towns. They found that certain age groups had increased risk of homicide with no outliers in any counties.

## Hypothesis


We predict that there will be a significant difference between the crime rates in the college cities/town and their local county, especially the property crimes. Through our background research, we found a study published in the American Journal of Police indicating that there were significantly higher property crime rates yet lower crime rates in the university area compared to the cities/county due to college students being easier targets, and the findings were consistent with other studies such as from University of South Florida. Our null hypothesis is there is no significant difference in the aforementioned above and our alternative hypothesis is that there is a significant difference in the aforementioned above.
