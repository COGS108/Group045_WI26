## Background and Prior Work

Crime in college and university towns has been widely reported for the last few decades, with government reported statistics going back to the 1990s. The <a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Jeanne Clery Disclosure of Campus Security Policy and Campus Crime Statistics Act requires that crime data for universities must be publicly reported through a Department of Education sponsored survey. There is an abundance of data on isolated college crimes, but less so on the crime rates of areas near universities. <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)A study in the American Journal of Police compared campus and city crime rates, published in 1995 by Max L. Bromley from the University of South Florida, examined the relationship between the crime rates with the local communities that surround the university campus. The data used in the study relied heavily on the two largest universities in each state and on an average of 2 years for the population. In result, there was a similar widespread variation found when comparing the violent and property crime rates per 1,000 students or city residents. All of the comparisons concluded that the university's crime rates were significantly lower than those of their cities. In addition, there was a finding that property crime was the vast majority of crimes compared to violent crimes, and that the university property crimes were especially higher than those in their cities. Overall, the findings were consistent with the prior studies in the 1980s and the early 1990s. <a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3)A research paper published on November 24 2020 by Kelsey Cundiff at University of Missouri-St. Louis analyzed campus proximity and neighborhood crime rates. It hypothesized with opportunity theory that neighborhoods within close proximity to college campuses will have higher crime rates because “college students are vulnerable to victimisation" due to convergence for their social lifestyles. It used data collected by the National Neighborhood Crime Survey, and analysis was done with a “multilevel negative binomial regression” to analyze the relationship between “proximity… to college campus[es] and rates of violent and property crime”. The researchers doing this study found that “bordering tracts higher rates of larceny, burglary, and robbery,” acting as proxies for “indicators of crime.” <a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) A research study published on October 10, 2023 by Carter Prentice and his team conducted a study on whether there was a difference in homicide and suicide rates between college towns/cities vs non college towns. Prentice’s team conducted this study by analyzing county level mortality rates from the CDC death certificate data from the years 2015 to 2019. To determine if a county was deemed a college town they used the classifications of the American Communities Project (ACP) to determine whether it was a college town or not. In college towns, they found that there was a median 2.56 total homicides per 100000 population and in non college towns, a 3.11 total homicide median. But in college towns, they found that people aged less than 24 had a higher rate of firearm homicide compared to non college towns, showing elevated risk in younger people with firearms. The researchers found that the results for population level homicide were not significantly different from college towns and non college towns. They found that certain age groups had increased risk of homicide with no outliers in any counties.

There are still notable gaps in the existing research, since studies focus on the close proximity of crime near colleges, or on certain categories of crime(either property or violent crime). Additionally, these studies are fairly old, with the 2020 study at the University of Missouri St. Louis being the most recent to analyse campus proximity in relation to crime.  Additionally, the study conducted in 1995 was conducted for the South Florida region, so we hope to investigate the findings in a different region (California).

### FBI NIBRS California 2024 —> County Offense Counts & Rates


### Describing the important metrics:
This dataset comes from the FBI’s NIBRS incident-based reporting for California in 2024. After joining the offense table with offense type and joining incidents to agencies/counties, each offense record can be counted and summarized by county. Our important metrics are: offense_rows, offense_name and the offense_category_name it fits into like Aggrevated assult would be categorized as a violent crime, and offences_per_100k.

offense_rows: The total number of reported offense records in a county for a given offense type. This represents the crimes reported to law enforcement agencies in each county. For example, if Los Angeles has 5,000 theft offenses, that means 5,000 separate theft incidents were officially recorded by law enforcement agencies in that county during 2024.

offenses_per_100k: This metric helps standarde the offense counts per 100,000 residents using county population data. The rate per 100,000 helps compare counties of different sizes more fairly than raw counts. For instance, a small county with 50,000 people reporting 100 burglaries has a rate of 200 per 100k, which can be directly compared to a large county with 1 million people reporting 3,000 burglaries, so 300 per 100k.

offense_name and offense_category_name: These classify crimes into specific types and broader categories like "Violent Crimes".


### Major Concerns:
There are a few notable concerns that arise from the dataset. Firstly, not all law enforcement agencies report to NIBRS, with the five most populated states (including California) having a combined participation rate of 48% ([Congress.gov](www.congress.gov/crs-product/IF13087)). This combined with many college campuses having their own police forces can lead to underreporting and some student crimes not making any appearances in the county-level data, lowering college town crime rates. Contrastingly, some crimes may be double reported to both the county-level police and campus police. Additionally, there may be behavioral differences in the reporting of college students as compared to permanent residents. Students may be less likely to report certain smaller property crimes than a permanent resident who has more at stake to NOT report. Students may also be discouraged by the tedious process and therefore would fail to report. Next, police presence may vary from town to town, with college towns often having higher presence, leading to higher detection. Furthermore, there are some geographical concerns. Different agencies through the state may classify similar crimes differently, changing the proportion of certain crimes to others. Also, differentiating the boundaries of a college town can be tricky, students live both on campus and around which can create a haziness between resident and student. Lastly, the data coming in annual reports poses the challenge of student population fluctuation, the crime rates and populations could be vastly different during school season as opposed to breaks like summer and winter. 


“Federal Support for Law Enforcement Agencies’ Transition to the National Incident-Based Reporting System (NIBRS).” Congress.gov, 2025, www.congress.gov/crs-product/IF13087.


```python
# YOUR CODE TO LOAD/CLEAN/TIDY/WRANGLE THE DATA GOES HERE
import pandas as pd
import os

DATA_DIR = "data/00-raw/CA-2024"
OUT_DIR = "data/02-processed"
os.makedirs(OUT_DIR, exist_ok=True)

ag  = pd.read_csv(f"{DATA_DIR}/agencies.csv", encoding="latin1", low_memory=False)
off = pd.read_csv(f"{DATA_DIR}/NIBRS_OFFENSE.csv", low_memory=False)
ot  = pd.read_csv(f"{DATA_DIR}/NIBRS_OFFENSE_TYPE.csv", low_memory=False)
inc = pd.read_csv(f"{DATA_DIR}/NIBRS_incident.csv", low_memory=False)

# filter to CA + 2024
ag = ag[
    (ag["data_year"] == 2024) &
    (ag["state_postal_abbr"] == "CA") &
    (ag["publishable_flag"] == "Y")
]
off = off[off["data_year"] == 2024]
inc = inc[inc["data_year"] == 2024]

# county population
county_pop = (ag.groupby("county_name", as_index=False)["population"]
                .sum()
                .rename(columns={"population": "county_population"}))

# incident -> agency -> county
inc_ag = inc[["incident_id", "agency_id", "data_year"]].merge(
    ag[["agency_id", "data_year", "county_name"]],
    on=["agency_id", "data_year"],
    how="inner"
)

# offense -> offense name/category -> county
off_named = off.merge(
    ot[["offense_code", "offense_name", "offense_category_name"]],
    on="offense_code",
    how="left"
).merge(
    inc_ag[["incident_id", "county_name"]],
    on="incident_id",
    how="inner"
)

missing = off_named.isna().mean().sort_values(ascending=False)
display(missing.head(15))

# counts + rates
counts = (off_named.groupby(["county_name", "offense_category_name", "offense_name"], as_index=False)
                  .size()
                  .rename(columns={"size": "offense_rows"}))

final = counts.merge(county_pop, on="county_name", how="left")
final["offenses_per_100k"] = final["offense_rows"] / final["county_population"] * 100000

display(final.sort_values("offense_rows", ascending=False).head(20))

final.to_csv(f"{OUT_DIR}/ca2024_county_offense_rates.csv", index=False)
county_pop.to_csv(f"{OUT_DIR}/ca2024_county_population.csv", index=False)

print("Saved:",
      f"{OUT_DIR}/ca2024_county_offense_rates.csv",
      "and",
      f"{OUT_DIR}/ca2024_county_population.csv")
```
